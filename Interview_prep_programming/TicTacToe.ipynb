This code will create a game of tic tac toe with random moves by human as well as computer. If we want both players to play strategically we will have to apply Reinforcement Learning in this.

```python
import random
class TicTacToe:
    def __init__(self):
        self.board_size = 3
        self.filled_pos = set()
        self.turn = "human"
        self.human_val_choice = None
        self.board = None
        
    def set_board(self):
        list_print = [[' ', "|", " ", "|", " "], ['-', "+", "-", "+", "-"], [' ', "|", " ", "|", " "], ['-', "+", "-", "+", "-"], [' ', "|", " ", "|", " "]]
        self.board = list_print
        
    def print_board(self):
        for i in range(len(self.board)):
            print("".join(self.board[i]))
            
    def get_input_pos(self):
        pos = input("enter your placement position(1-9):")
        if pos in self.filled_pos:
            pos = input("That position is taken!Enter your placement position(1-9):")
        return pos
    
    def get_input_val(self):
        pos = input("enter your placement value[0 or X]:")
        self.human_val_choice = pos
        return pos
    
    def fill_value(self, pos, val):
        self.filled_pos.add(pos)
        mapping = {'1':(0,0), '2': (0, 2), '3':(0, 4), '4': (2, 0), '5': (2, 2), '6': (2, 4), '7': (4, 0), '8': (4, 2), '9': (4, 4)}
        i, j = mapping[pos]
        self.board[i][j] = val.upper()
         
    def is_game_over(self):
        return len(self.filled_pos) == 9
        
    def is_it_win(self):
        diag_0 = (self.board[0][0] == '0' and self.board[2][2] == '0' and self.board[4][4] == '0') 
        diag_x = (self.board[0][0] == 'X' and self.board[2][2] == 'X' and self.board[4][4] == 'X')
        col_0_0 = (self.board[0][0] == '0' and self.board[2][0] == '0' and self.board[4][0] == '0')
        col_0_x = (self.board[0][0] == 'X' and self.board[2][0] == 'X' and self.board[4][0] == 'X')
        col_1_0 = (self.board[0][2] == '0' and self.board[2][2] == '0' and self.board[4][2] == '0')
        col_1_x = (self.board[0][2] == 'X' and self.board[2][2] == 'X' and self.board[4][2] == 'X')
        col_2_0 = (self.board[0][4] == '0' and self.board[2][4] == '0' and self.board[4][4] == '0')
        col_2_x = (self.board[0][4] == 'X' and self.board[2][4] == 'X' and self.board[4][4] == 'X')
        row_0_0 = (self.board[0][0] == '0' and self.board[0][2] == '0' and self.board[0][4] == '0')
        row_0_x = (self.board[0][0] == 'X' and self.board[0][2] == 'X' and self.board[0][4] == 'X')
        row_1_0 = (self.board[2][0] == '0' and self.board[2][2] == '0' and self.board[2][4] == '0')
        row_1_x = (self.board[2][0] == 'X' and self.board[2][2] == 'X' and self.board[2][4] == 'X')
        row_2_0 = (self.board[4][0] == '0' and self.board[4][2] == '0' and self.board[4][4] == '0')
        row_2_x = (self.board[4][0] == 'X' and self.board[4][2] == 'X' and self.board[4][4] == 'X')
#         print(diag_0, diag_x)
#         print(col_0_0, col_0_x, col_1_0, col_1_x, col_2_0, col_2_x)
#         print(row_0_0, row_0_x, row_1_0, row_1_x, row_2_0, row_2_x)
        return (diag_0 or diag_x or col_0_0 or col_0_x or col_1_0 or col_1_x or col_2_0 or col_2_x or row_0_0 or row_0_x or row_1_0 or row_1_x or row_2_0 or row_2_x)
    
    def computer_turn(self):
        rand_choice_val = "X" if self.human_val_choice is '0' else '0'
        positions = range(1, 10)
        positions_l = [str(i) for i in positions if str(i) not in self.filled_pos]
        print("positions eligible:", positions_l)
        rand_choice_pos = random.choice(positions_l)
        self.fill_value(rand_choice_pos, rand_choice_val)
        return rand_choice_pos
                        
def play_game():
    game = TicTacToe()
    game.set_board()
    game.print_board()
    val = game.get_input_val()
    while True:
        if game.turn == "human": 
            print("It's your turn")
            pos = game.get_input_pos()
            print("pos:", pos)
            game.fill_value(pos, game.human_val_choice)
            game.print_board()
            game.turn = "computer"
        elif game.turn == "computer": 
            print("It's computer's turn")
            pos = game.computer_turn()
            print("pos:", pos)
            game.print_board()
            game.turn = "human"         
        if game.is_it_win():
            if game.turn == "human":
                print("Oops, You lost!")
                break
            else:
                print("You Won!!")
                break
        if game.is_game_over():
            print("Game draw!")
            break
        
```

```python
play_game()
```
